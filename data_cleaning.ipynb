```python
# add missing flags
df_full['BuildingArea_missing'] = df_full['BuildingArea'].isna().astype(int)
df_full['Car_missing'] = df_full['Car'].isna().astype(int)
df_full['Bathroom_missing'] = df_full['Bathroom'].isna().astype(int)
df_full['Bedroom2_missing'] = df_full['Bedroom2'].isna().astype(int)
df_full['Landsize_missing'] = df_full['Landsize'].isna().astype(int)
df_full['Lattitude_missing'] = df_full['Lattitude'].isna().astype(int)
df_full['Longtitude_missing'] = df_full['Longtitude'].isna().astype(int)
print(df_full.columns)
```

```python
# Data Cleaning and Preparation for Melbourne Housing Dataset
import numpy as np

def clean_data(df_full):
    
    # Drop rows with missing data in column: 'Price'
    df_full = df_full.dropna(subset=['Price'])
    
    # Add a new column with the log of 'Price'
    df_full['Log_Price'] = np.log(df_full['Price'])
    df_full['Lattitude'] = df_full.groupby('Suburb')['Lattitude'].transform(lambda x: x.fillna(x.median()))
    df_full['Longtitude'] = df_full.groupby('Suburb')['Longtitude'].transform(lambda x: x.fillna(x.median()))
    
    # Drop column: 'YearBuilt'
    df_full = df_full.drop(columns=['YearBuilt'])
    
    # impute building_area
    # add dummy to keep track of imputed rows
    df_full['BuildingArea_missing'] = df_full['BuildingArea'].isna().astype(int)
    # impute buildingarea by suburb, rooms, car spots
    df_full['BuildingArea'] = df_full.groupby(['Suburb', 'Rooms', 'Car'])['BuildingArea'].transform(lambda x: x.fillna(x.median()))
    # Fallback 1: Suburb + Rooms
    df_full['BuildingArea'] = df_full.groupby(['Suburb', 'Rooms'])['BuildingArea'].transform(lambda x: x.fillna(x.median()))
    # Fallback 2: Rooms only
    df_full['BuildingArea'] = df_full.groupby('Rooms')['BuildingArea'].transform(lambda x: x.fillna(x.median()))
    
    # impute car bedroom2 and bathroom by rooms  and suburb
    cols = ['Car', 'Bathroom', 'Bedroom2']
    for col in cols:
        df_full[col] = df_full.groupby(['Suburb', 'Rooms'])[col].transform(lambda x: x.fillna(x.median()))
    
    # impute landsize by suburb and type
    #df_full['Landsize'] = df_full.groupby(['Suburb', 'Type'])['Landsize'].transform(lambda x: x.fillna(x.median()))
    df_full['Landsize'] = df_full.groupby('Suburb')['Landsize'].transform(lambda x: x.fillna(x.median()))
    return df_full

# df_full_clean = clean_data(df_full.copy())
# df_full_clean.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = [
    ('BuildingArea', 'BuildingArea_missing'),
    ('Car', 'Car_missing'),
    ('Bathroom', 'Bathroom_missing'),
    ('Bedroom2', 'Bedroom2_missing'),
    ('Landsize', 'Landsize_missing'),
    ('Lattitude', 'Lattitude_missing'),
    ('Longtitude', 'Longtitude_missing')
]

for col, flag in cols:
    print(f"\n=== {col} ===")
    print("Non-imputed summary:")
    print(df_full[df_full[flag] == 0][col].describe())
    print("Imputed summary:")
    print(df_full[df_full[flag] == 1][col].describe())
    
    # Exclude outliers for BuildingArea and Landsize using 1st and 99th percentiles
    if col in ['BuildingArea', 'Landsize']:
        lower = df_full[col].quantile(0.01)
        upper = df_full[col].quantile(0.99)
        plot_data = df_full[(df_full[col] >= lower) & (df_full[col] <= upper)]
    else:
        plot_data = df_full

    plt.figure(figsize=(10,4))
    sns.histplot(
        data=plot_data, x=col, hue=flag, 
        bins=30, kde=True, stat='density', 
        palette={0: "blue", 1: "red"}, alpha=0.5
    )
    plt.title(f'Distribution of {col}: Imputed (red) vs Non-imputed (blue)')
    plt.legend(title='Imputed', labels=['Non-imputed', 'Imputed'])
    plt.show()
```
